### 2.1. Loading the Data

```python
# Function to load images and masks
def load_data(image_dir, mask_dir, img_height, img_width,color_mode="rgb"):
    image_paths = sorted(os.listdir(image_dir))  # Sort for matching pairs
    mask_paths = sorted(os.listdir(mask_dir))

    images = []
    masks = []

    for i in range(len(image_paths)):
        # Load and preprocess image
        img_path = os.path.join(image_dir, image_paths[i])
        image = load_img(img_path, target_size=(img_height, img_width))  # Resize
        image = img_to_array(image) / 255.0  # Normalize to [0, 1]
        images.append(image)

        # Load and preprocess mask
        mask_path = os.path.join(mask_dir, mask_paths[i])
        mask = load_img(mask_path, target_size=(img_height, img_width),color_mode=color_mode)  # Resize
        mask = img_to_array(mask) / 255.0 # Normalize mask to [0, 1] (binary)
        masks.append(mask)

    return np.array(images), np.array(masks)
```

```python
@tf.function
def dice_coefficient(y_true, y_pred, smooth=1e-6):
    intersection = K.sum(K.abs(y_true * y_pred), axis=-1)
    union = K.sum(K.abs(y_true) + K.abs(y_pred), axis=-1)
    return (2. * intersection + smooth) / (union + smooth)

@tf.function
def iou_metric(y_true, y_pred, smooth=1e-6):
    intersection = K.sum(K.abs(y_true * y_pred), axis=-1)
    union = K.sum(K.abs(y_true) + K.abs(y_pred), axis=-1) - intersection
    return (intersection + smooth) / (union + smooth)

@tf.function
def dice_loss(y_true, y_pred, smooth=1e-6):
    intersection = K.sum(K.abs(y_true * y_pred), axis=-1)
    union = K.sum(K.abs(y_true) + K.abs(y_pred), axis=-1)
    return 1 - (2. * intersection + smooth) / (union + smooth)
```

```python
#LeakyRelu Unet with padding
def unet(input_size=(256, 256, 3), num_classes=3):
    inputs = layers.Input(input_size)

    # Contracting Path (Encoder)
    c1 = layers.Conv2D(64, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(inputs)
    c1 = layers.Conv2D(64, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(c1)
    p1 = layers.MaxPooling2D((2, 2))(c1)

    c2 = layers.Conv2D(128, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(p1)
    c2 = layers.Conv2D(128, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(c2)
    p2 = layers.MaxPooling2D((2, 2))(c2)

    c3 = layers.Conv2D(256, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(p2)
    c3 = layers.Conv2D(256, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(c3)
    p3 = layers.MaxPooling2D((2, 2))(c3)

    c4 = layers.Conv2D(512, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(p3)
    c4 = layers.Conv2D(512, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(c4)
    p4 = layers.MaxPooling2D((2, 2))(c4)

    c5 = layers.Conv2D(1024, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(p4)
    c5 = layers.Conv2D(1024, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(c5)

    # Expansive Path (Decoder)
    u6 = layers.Conv2DTranspose(512, (3, 3), strides=(2, 2), padding='same')(c5)
    u6 = layers.concatenate([u6, c4])
    c6 = layers.Conv2D(512, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(u6)
    c6 = layers.Conv2D(512, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(c6)

    u7 = layers.Conv2DTranspose(256, (3, 3), strides=(2, 2), padding='same')(c6)
    u7 = layers.concatenate([u7, c3])
    c7 = layers.Conv2D(256, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(u7)
    c7 = layers.Conv2D(256, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(c7)

    u8 = layers.Conv2DTranspose(128, (3, 3), strides=(2, 2), padding='same')(c7)
    u8 = layers.concatenate([u8, c2])
    c8 = layers.Conv2D(128, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(u8)
    c8 = layers.Conv2D(128, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(c8)

    u9 = layers.Conv2DTranspose(64, (3, 3), strides=(2, 2), padding='same')(c8)
    u9 = layers.concatenate([u9, c1])
    c9 = layers.Conv2D(64, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(u9)
    c9 = layers.Conv2D(64, (3, 3), activation=LeakyReLU(alpha=0.1), padding='same')(c9)

    # Output Layer
    outputs = layers.Conv2D(num_classes, (1, 1), activation='softmax')(c9)  # For multi-class segmentation

    model = models.Model(inputs=[inputs], outputs=[outputs])

    return model
    
# Create the U-Net model
model = unet(input_size=(IMG_HEIGHT, IMG_WIDTH, 3), num_classes=3)

optimizer=tf.keras.optimizers.Adam(learning_rate=0.000006, clipnorm=1.0)
# Compile the model
model.compile(optimizer=optimizer,
              loss='categorical_crossentropy',  # For multi-class segmentation
              metrics=['accuracy', MeanIoU(num_classes=3)])  # Pass num_classes to MeanIoU

print("PASS")
```

```python
#Customized Unet with less filters
import tensorflow as tf
from tensorflow.keras import layers, models

def custom_unet(input_size=(256, 256, 3), num_classes=3):
    inputs = layers.Input(input_size)

    # Encoder (Downsampling Path)
    c1 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    c1 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(c1)
    p1 = layers.MaxPooling2D((2, 2))(c1)

    c2 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(p1)
    c2 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c2)
    p2 = layers.MaxPooling2D((2, 2))(c2)

    c3 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(p2)
    c3 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c3)
    p3 = layers.MaxPooling2D((2, 2))(c3)

    c4 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(p3)
    c4 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(c4)
    p4 = layers.MaxPooling2D((2, 2))(c4)

    # Bottleneck
    c5 = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(p4)
    c5 = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(c5)

    # Decoder (Upsampling Path)
    u6 = layers.Conv2DTranspose(256, (3, 3), strides=(2, 2), padding='same')(c5)
    u6 = layers.concatenate([u6, c4])
    c6 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(u6)
    c6 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(c6)

    u7 = layers.Conv2DTranspose(128, (3, 3), strides=(2, 2), padding='same')(c6)
    u7 = layers.concatenate([u7, c3])
    c7 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(u7)
    c7 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c7)

    u8 = layers.Conv2DTranspose(64, (3, 3), strides=(2, 2), padding='same')(c7)
    u8 = layers.concatenate([u8, c2])
    c8 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(u8)
    c8 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c8)

    u9 = layers.Conv2DTranspose(32, (3, 3), strides=(2, 2), padding='same')(c8)
    u9 = layers.concatenate([u9, c1])
    c9 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(u9)
    c9 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(c9)

    # Output Layer
    outputs = layers.Conv2D(num_classes, (1, 1), activation='softmax')(c9)  # For multi-class segmentation

    model = models.Model(inputs=[inputs], outputs=[outputs])

    return model

# Instantiate and compile the model
model = custom_unet(input_size=(256, 256, 3), num_classes=3)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()
```

```python
def make_unet_experimental_fixed():
    input_size = (256, 256, 3)
    inputs = Input(input_size)

    # Encoder without pool1, pool2, pool3
    conv1 = Conv2D(8, 3, activation='relu', padding='same', kernel_initializer='he_normal')(inputs)
    conv1 = Conv2D(8, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv1)

    conv2 = Conv2D(16, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv1)
    conv2 = Conv2D(16, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv2)

    conv3 = Conv2D(32, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv2)
    conv3 = Conv2D(32, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv3)

    # Single pooling layer (pool4)
    conv4 = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv3)
    conv4 = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv4)
    pool4 = MaxPooling2D(pool_size=(2, 2),strides=(2,2))(conv4)

    # Bottleneck
    conv5 = Conv2D(128, 3, activation='relu', padding='same', kernel_initializer='he_normal')(pool4)
    conv5 = Conv2D(128, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv5)

    # Decoder with UpSampling and skip connections
    up6 = Conv2D(64, 2, activation='relu', padding='same', kernel_initializer='he_normal')(UpSampling2D(size=(2, 2))(conv5))
    merge6 = concatenate([conv4, up6], axis=3)  # Shape alignment is correct here
    conv6 = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(merge6)
    conv6 = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv6)

    # Adjust conv3 to match up7's size
    conv3_resized = Conv2D(32, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv3)
    merge7 = concatenate([conv3_resized, conv6], axis=3)  # Shapes now align
    conv7 = Conv2D(32, 3, activation='relu', padding='same', kernel_initializer='he_normal')(merge7)
    conv7 = Conv2D(32, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv7)

    # Adjust conv2 to match up8's size
    conv2_resized = Conv2D(16, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv2)
    merge8 = concatenate([conv2_resized, conv7], axis=3)  # Shapes now align
    conv8 = Conv2D(16, 3, activation='relu', padding='same', kernel_initializer='he_normal')(merge8)
    conv8 = Conv2D(16, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv8)

    # Adjust conv1 to match up9's size
    conv1_resized = Conv2D(8, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv1)
    merge9 = concatenate([conv1_resized, conv8], axis=3)  # Shapes now align
    conv9 = Conv2D(8, 3, activation='relu', padding='same', kernel_initializer='he_normal')(merge9)
    conv9 = Conv2D(8, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv9)

    conv10 = Conv2D(3, 1, activation='softmax')(conv9)

    model = Model(inputs=inputs, outputs=conv10)
    return model

# Compile the model
model = make_unet_experimental_fixed()

model.compile(
    optimizer=Adam(learning_rate=0.00001),
    loss=dice_loss,
    metrics=['accuracy',dice_coefficient, iou_metric]
)
print("Model Ready")
```

```python
#Enhanced code of V1 model

def make_unet_advanced():
    input_size = (256, 256, 3)
    inputs = Input(input_size)

    # Encoding path
    conv1 = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(inputs)
    conv1 = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv1)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

    conv2 = Conv2D(128, 3, activation='relu', padding='same', kernel_initializer='he_normal')(pool1)
    conv2 = Conv2D(128, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv2)
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

    conv3 = Conv2D(256, 3, activation='relu', padding='same', kernel_initializer='he_normal')(pool2)
    conv3 = Conv2D(256, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv3)
    pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)

    conv4 = Conv2D(512, 3, activation='relu', padding='same', kernel_initializer='he_normal')(pool3)
    conv4 = Conv2D(512, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv4)
    drop4 = Dropout(0.5)(conv4)
    pool4 = MaxPooling2D(pool_size=(2, 2))(drop4)

    conv5 = Conv2D(1024, 3, activation='relu', padding='same', kernel_initializer='he_normal')(pool4)
    conv5 = Conv2D(1024, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv5)
    drop5 = Dropout(0.5)(conv5)

    # Decoding path
    up6 = Conv2D(512, 2, activation='relu', padding='same', kernel_initializer='he_normal')(UpSampling2D(size=(2, 2))(drop5))
    merge6 = concatenate([drop4, up6], axis=3)
    conv6 = Conv2D(512, 3, activation='relu', padding='same', kernel_initializer='he_normal')(merge6)
    conv6 = Conv2D(512, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv6)

    up7 = Conv2D(256, 2, activation='relu', padding='same', kernel_initializer='he_normal')(UpSampling2D(size=(2, 2))(conv6))
    merge7 = concatenate([conv3, up7], axis=3)
    conv7 = Conv2D(256, 3, activation='relu', padding='same', kernel_initializer='he_normal')(merge7)
    conv7 = Conv2D(256, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv7)

    up8 = Conv2D(128, 2, activation='relu', padding='same', kernel_initializer='he_normal')(UpSampling2D(size=(2, 2))(conv7))
    merge8 = concatenate([conv2, up8], axis=3)
    conv8 = Conv2D(128, 3, activation='relu', padding='same', kernel_initializer='he_normal')(merge8)
    conv8 = Conv2D(128, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv8)

    up9 = Conv2D(64, 2, activation='relu', padding='same', kernel_initializer='he_normal')(UpSampling2D(size=(2, 2))(conv8))
    merge9 = concatenate([conv1, up9], axis=3)
    conv9 = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(merge9)
    conv9 = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv9)

    conv10 = Conv2D(3, 1, activation='softmax')(conv9)

    model = Model(inputs=inputs, outputs=conv10)
    return model

model = make_unet_advanced()

# Compile with an adjusted learning rate
model.compile(
    optimizer=Adam(learning_rate=0.0000005),
    loss='categorical_crossentropy',
    metrics=['accuracy',MeanIoU(num_classes=3)]
)
```

```python
#Janani's architecture

def UNet(input_shape, IMG_CLASSES):
    inputs = Input(input_shape) 

    # (256, 256, 3)
    # First convolutional block
    c1 = Conv2D(16, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(inputs)
    c1 = Dropout(0.1)(c1)
    c1 = Conv2D(16, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c1)
    p1 = MaxPooling2D((2,2))(c1)  # Retain MaxPooling here

    #(128, 128, 3)
    # Second convolutional block
    c2 = Conv2D(32, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(p1)
    c2 = Dropout(0.1)(c2)
    c2 = Conv2D(32, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c2)

    #(128, 128, 3)
    # Third convolutional block
    c3 = Conv2D(64, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c2)
    c3 = Dropout(0.2)(c3)
    c3 = Conv2D(64, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c3)

    #(128, 128, 3)
    # Fourth convolutional block
    c4 = Conv2D(128, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c3)
    c4 = Dropout(0.2)(c4)
    c4 = Conv2D(128, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c4)

    #(128, 128, 3)
    # Bottleneck
    c5 = Conv2D(256, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c4)
    c5 = Dropout(0.3)(c5)
    c5 = Conv2D(256, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c5)

    #(128, 128, 3)
    # Decoder: Transpose convolutions with skip connections
    c6 = Conv2D(128, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c5)
    c6 = concatenate([c6, c4])
    c6 = Dropout(0.2)(c6)
    c6 = Conv2D(128, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c6)

    c7 = Conv2D(64, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c6)
    c7 = concatenate([c7, c3])
    c7 = Dropout(0.2)(c7)
    c7 = Conv2D(64, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c7)

    c8 = Conv2D(32, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c7)
    c8 = concatenate([c8, c2])
    c8 = Dropout(0.1)(c8)
    c8 = Conv2D(32, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c8)

    u9 = Conv2DTranspose(16, (2,2), strides=(2,2), padding="same")(c8)
    u9 = concatenate([u9, c1], axis=3)
    c9 = Conv2D(16, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(u9)
    c9 = Dropout(0.1)(c9)
    c9 = Conv2D(16, (3,3), activation="relu", kernel_initializer="he_normal", padding="same")(c9)

    # Output layer
    outputs = Conv2D(IMG_CLASSES, (1,1), activation="softmax")(c9)

    model = Model(inputs, outputs)
    return model

model = UNet((IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS),3)
model.compile(
    optimizer=Adam(learning_rate=0.000005),
    loss='categorical_crossentropy',
    metrics=['accuracy', MeanIoU(num_classes=3)]
)
print("Model Ready")
```

```python
#WNET

def make_wnet(input_size=(256, 256, 3), num_classes=3):
    inputs = Input(input_size)
    conv1 = Conv2D(16, 3, activation='relu', padding='same', kernel_initializer='he_normal')(inputs)
    conv1 = Conv2D(16, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv1)
    conv2 = Conv2D(32, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv1)
    conv2 = Conv2D(32, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv2)
    conv3 = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv2)
    conv3 = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv3)
    conv4 = Conv2D(128, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv3)
    conv4 = Conv2D(128, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv4)
    pool4 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(conv4)

    bottleneck = Conv2D(256, 3, activation='relu', padding='same', kernel_initializer='he_normal')(pool4)
    bottleneck = Conv2D(256, 3, activation='relu', padding='same', kernel_initializer='he_normal')(bottleneck)

    up6 = Conv2D(128, 2, activation='relu', padding='same', kernel_initializer='he_normal')(UpSampling2D(size=(2, 2))(bottleneck))
    merge6 = concatenate([conv4, up6], axis=3)
    conv6 = Conv2D(128, 3, activation='relu', padding='same', kernel_initializer='he_normal')(merge6)
    conv6 = Conv2D(128, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv6)

    conv3_resized = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv3)
    merge7 = concatenate([conv3_resized, conv6], axis=3)
    conv7 = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(merge7)
    conv7 = Conv2D(64, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv7)

    conv2_resized = Conv2D(32, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv2)
    merge8 = concatenate([conv2_resized, conv7], axis=3)
    conv8 = Conv2D(32, 3, activation='relu', padding='same', kernel_initializer='he_normal')(merge8)
    conv8 = Conv2D(32, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv8)

    conv1_resized = Conv2D(16, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv1)
    merge9 = concatenate([conv1_resized, conv8], axis=3)
    conv9 = Conv2D(16, 3, activation='relu', padding='same', kernel_initializer='he_normal')(merge9)
    conv9 = Conv2D(16, 3, activation='relu', padding='same', kernel_initializer='he_normal')(conv9)

    output = Conv2D(num_classes, 1, activation='softmax')(conv9)
       
    model = Model(inputs=inputs, outputs=output)
    return model

# Compile the model
model = make_wnet()

model.compile(
    optimizer=Adam(learning_rate=0.000005),
    loss='categorical_crossentropy',
    metrics=['accuracy', MeanIoU(num_classes=3)]
)

print("U-Net Model Ready")
```

```python
#DeepLab
def ASPP(inputs):
    shape = inputs.shape

    y_pool = AveragePooling2D(pool_size=(shape[1], shape[2]), name='average_pooling')(inputs)
    y_pool = Conv2D(filters=256, kernel_size=1, padding='same', use_bias=False)(y_pool)
    y_pool = BatchNormalization(name=f'bn_1')(y_pool)
    y_pool = Activation('relu', name=f'relu_1')(y_pool)
    y_pool = UpSampling2D((shape[1], shape[2]), interpolation="bilinear")(y_pool)

    y_1 = Conv2D(filters=256, kernel_size=1, dilation_rate=1, padding='same', use_bias=False)(inputs)
    y_1 = BatchNormalization()(y_1)
    y_1 = Activation('relu')(y_1)

    y_6 = Conv2D(filters=256, kernel_size=3, dilation_rate=6, padding='same', use_bias=False)(inputs)
    y_6 = BatchNormalization()(y_6)
    y_6 = Activation('relu')(y_6)

    y_12 = Conv2D(filters=256, kernel_size=3, dilation_rate=12, padding='same', use_bias=False)(inputs)
    y_12 = BatchNormalization()(y_12)
    y_12 = Activation('relu')(y_12)

    y_18 = Conv2D(filters=256, kernel_size=3, dilation_rate=18, padding='same', use_bias=False)(inputs)
    y_18 = BatchNormalization()(y_18)
    y_18 = Activation('relu')(y_18)

    y = Concatenate()([y_pool, y_1, y_6, y_12, y_18])

    y = Conv2D(filters=256, kernel_size=1, dilation_rate=1, padding='same', use_bias=False)(y)
    y = BatchNormalization()(y)
    y = Activation('relu')(y)
    return y


def DeepLabV3Plus(input_shape, IMG_CLASSES):
    # Inputs
    inputs = layers.Input(input_shape)

    # Pre-trained ResNet50
    base_model = ResNet50(weights='imagenet', include_top=False, input_tensor=inputs)

    # Pre-trained ResNet50 Output
    image_features = base_model.get_layer('conv4_block6_out').output
    x_a = ASPP(image_features)
    x_a = UpSampling2D((4, 4), interpolation="bilinear")(x_a)

    # Get low-level features
    x_b = base_model.get_layer('conv2_block2_out').output
    x_b = Conv2D(filters=48, kernel_size=1, padding='same', use_bias=False)(x_b)
    x_b = BatchNormalization()(x_b)
    x_b = Activation('relu')(x_b)

    x = Concatenate()([x_a, x_b])

    x = Conv2D(filters=256, kernel_size=3, padding='same', activation='relu',use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(filters=256, kernel_size=3, padding='same', activation='relu', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = UpSampling2D((4, 4), interpolation="bilinear")(x)

    # Outputs
    x = Conv2D(IMG_CLASSES, (1, 1), name='output_layer')(x)
    x = Activation('softmax')(x)

    # Model
    model = Model(inputs=inputs, outputs=x)
    return model

model = DeepLabV3Plus((IMG_HEIGHT,IMG_WIDTH,IMG_CHANNELS),3)
model.compile(optimizer='adam', loss=total_loss, metrics = ["accuracy"])
print("DeepLab Model Ready")
```

```python
input_shape = (256, 256, 3)  # 256x256 image with 3 channels (RGB)
num_classes = 3  # 3 classes for segmentation



# Build the model
def build_resnet_segmentation_model(input_shape, num_classes):
    # Input layer
    inputs = layers.Input(shape=input_shape)
    # ResNet50 as the backbone (encoder) with no top layers (include_top=False)
    resnet = ResNet50(include_top=False, input_shape=input_shape)

    # Freeze ResNet layers
    resnet.trainable = True
    
    # Encoder - ResNet layers
    x = resnet(inputs)
    
    # Decoder: Upsample the feature maps

    # Upsampling layers

    x = UpSampling2D(size=(4, 4))(x)  # Upsample by 2x
    x = Conv2D(256, kernel_size=(3, 3), padding='same', activation='relu', kernel_initializer="he_normal")(x)
    x = UpSampling2D(size=(4, 4))(x)  # Upsample by 2x
    x = Conv2D(128, kernel_size=(3, 3), padding='same', activation='relu', kernel_initializer="he_normal")(x)
    x = UpSampling2D(size=(2, 2))(x)  # Upsample by 2x
    x = Conv2D(64, kernel_size=(3, 3), padding='same', activation='relu', kernel_initializer="he_normal")(x)
    x = Conv2D(32, kernel_size=(3, 3), padding='same', activation='relu', kernel_initializer="he_normal")(x)
    x = Conv2D(16, kernel_size=(3, 3), padding='same', activation='relu', kernel_initializer="he_normal")(x)
    
    # Output layer: 3 channels, SoftMax for multi-class segmentation
    outputs = Conv2D(num_classes, kernel_size=(1, 1), activation='softmax')(x)
    
    # Define the model
    model = models.Model(inputs, outputs)
    return model

# Build the model

model = build_resnet_segmentation_model(input_shape, num_classes)
# Compile the model with appropriate loss and metrics for segmentation

categorical_accuracy = tf.keras.metrics.CategoricalAccuracy()

# Compile the model with more appropriate metrics for segmentation
model.compile(optimizer=Adam(learning_rate=0.0001), 
              loss=dice_loss,  # Using Dice Loss for better handling of class imbalance
              metrics=[dice_coefficient, iou_metric, 'accuracy'])  # Adding Dice and IoU metrics

# Display the model architecture
model.summary()
```

## 5. Image Enhancement

```python
def dominant_color_mask(image):
    # Ensure the input is in the correct shape and range
    if image.dtype != np.float32 and image.dtype != np.float64:
        image = image.astype(np.float32)
    if image.max() > 1:  # Normalize to range [0, 1] if necessary
        image = image / 255.0

    # Create an empty array for the output image
    output = np.zeros_like(image)
    
    # Find the index of the dominant channel for each pixel
    dominant_channel = np.argmax(image, axis=-1)  # Shape: (H, W)
    
    # Map each dominant channel to its respective color
    output[dominant_channel == 0] = RED  # Red
    output[dominant_channel == 1] = GREEN  # Green
    output[dominant_channel == 2] = BLUE  # Blue

    # Scale the result back to the range [0, 255]    
    return output
```

### 8.1 MeanIOU per class

```python
def mean_iou_per_class(y_true, y_pred, num_classes):
    assert y_true.shape == y_pred.shape, "The shape of y_true and y_pred must be the same"
    
    # Initialize an array to store the IoU values for each class
    iou_per_class = np.zeros(num_classes)
    colors = [RED_COLOR, GREEN_COLOR, BLUE_COLOR]
    for cls in range(num_classes): # 0, 1, 2
        iou_sum = 0
        num_images = y_true.shape[0]
        color = colors[cls]
        image = np.full((IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), color, dtype=np.uint8)
        
        # Loop through each image
        for i in range(num_images): # 0, ... n
            true_class = ((y_true[i]*255) == color)  # Create a boolean mask for the true class
            pred_class = ((y_pred[i]*255) == color)  # Create a boolean mask for the predicted class
            
            # Calculate the intersection and union
            intersection = np.sum(true_class & pred_class)
            union = np.sum(true_class | pred_class)
            
            # IoU for the current image (avoid division by zero)
            iou = intersection / union if union != 0 else 0
            iou_sum += iou
            #print(f'Image no: {i}\t Class: {cls}\t iou: {iou}\t Union: {union}\t intersection: {intersection}')
        
        # Mean IoU for the current class across all images
        mean_iou = iou_sum / num_images
        iou_per_class[cls] = mean_iou

    # Print the Mean IoU for each class
    for cls in range(num_classes):
        print(f"Mean IoU for class {cls}: {iou_per_class[cls]:.4f}")
```

### 8.2 Overall MeanIOU

```python
def mean_iou(y_true, y_pred, num_classes=3):
    assert y_true.shape == y_pred.shape, "The shape of y_true and y_pred must be the same"
    num_images = y_true.shape[0]
    iou_sum=0
    # Loop through each image
    for i in range(num_images): # 0, ... n
        # Calculate the intersection and union
        intersection = np.sum(y_true[i] == y_pred[i])
        union = 2 * IMG_HEIGHT * IMG_WIDTH * IMG_CHANNELS - intersection
            
        # IoU for the current image (avoid division by zero)
        iou = intersection / union if union != 0 else 0
        iou_sum += iou
        #print(f'Image no: {i}\t Class: {cls}\t iou: {iou}\t Union: {union}\t intersection: {intersection}')
        
        # Mean IoU for the current class across all images
    mean_iou = iou_sum / num_images

    # Print the Mean IoU for each class
    print(f"Mean IoU : {mean_iou:.4f}")
```

### 8.3 Dice Coefficient per class

```python
def dice_coefficient_per_class(y_true, y_pred, num_classes):
    assert y_true.shape == y_pred.shape, "The shape of y_true and y_pred must be the same"
    
    # Initialize an array to store the IoU values for each class
    mean_dice_per_class = np.zeros(num_classes)
    colors = [RED_COLOR, GREEN_COLOR, BLUE_COLOR]
    for cls in range(num_classes): # 0, 1, 2
        dice_sum = 0
        num_images = y_true.shape[0]
        color = colors[cls]
        image = np.full((IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), color, dtype=np.uint8)
        
        # Loop through each image
        for i in range(num_images): # 0, ... n
            true_class = ((y_true[i]*255) == color)  # Create a boolean mask for the true class
            pred_class = ((y_pred[i]*255) == color)  # Create a boolean mask for the predicted class
            
            # Calculate the intersection and union
            intersection = 2 * np.sum(true_class & pred_class)
            union = 2 * IMG_HEIGHT * IMG_WIDTH * IMG_CHANNELS
            
            # IoU for the current image (avoid division by zero)
            dice = intersection / union
            dice_sum += dice
            #print(f'Image no: {i}\t Class: {cls}\t iou: {iou}\t Union: {union}\t intersection: {intersection}')
        
        # Mean IoU for the current class across all images
        mean_dice = dice_sum / num_images
        mean_dice_per_class[cls] = mean_dice

    # Print the Mean IoU for each class
    for cls in range(num_classes):
        print(f"Mean Dice for class {cls}: {mean_dice_per_class[cls]:.4f}")
```

### 8.4 Dice Coefficient Overall

```python
def dice_coefficient(y_true, y_pred, num_classes):
    assert y_true.shape == y_pred.shape, "The shape of y_true and y_pred must be the same"
    
    # Initialize an array to store the IoU values for each class    
    dice_sum = 0
    num_images = y_true.shape[0]
        
    # Loop through each image
    for i in range(num_images): # 0, ... n
        intersection = np.sum((y_true[i]*255) == (y_pred[i]*255)) * 2  # Create a boolean mask for the true class
            
        union = 2 * IMG_HEIGHT * IMG_WIDTH * IMG_CHANNELS
            
        # IoU for the current image (avoid division by zero)
        dice = intersection / union
        dice_sum += dice
        
    mean_dice = dice_sum / num_images

    # Print the Mean IoU for each class
    print(f"Mean Dice: {mean_dice:.4f}")
```

```python
def precision_multiclass(y_true, y_pred, num_classes):
    precision_scores = {}
    for i in range(num_classes):
        true_class = y_true[..., i].flatten()
        pred_class = y_pred[..., i].flatten()
        tp = np.sum(true_class * pred_class)
        pp = np.sum(pred_class)
        precision_scores[f'Class_{i}'] = tp / (pp + 1e-7)
    precision_scores['Mean_Precision'] = np.mean(list(precision_scores.values()))
    return precision_scores


def recall_multiclass(y_true, y_pred, num_classes):
    recall_scores = {}
    for i in range(num_classes):
        true_class = y_true[..., i].flatten()
        pred_class = y_pred[..., i].flatten()
        tp = np.sum(true_class * pred_class)
        ap = np.sum(true_class)
        recall_scores[f'Class_{i}'] = tp / (ap + 1e-7)
    recall_scores['Mean_Recall'] = np.mean(list(recall_scores.values()))
    return recall_scores


def f1_score_multiclass(y_true, y_pred, num_classes):
    f1_scores = {}
    for i in range(num_classes):
        precision = precision_multiclass(y_true, y_pred, num_classes)[f'Class_{i}']
        recall = recall_multiclass(y_true, y_pred, num_classes)[f'Class_{i}']
        f1_scores[f'Class_{i}'] = 2 * (precision * recall) / (precision + recall + 1e-7)
    f1_scores['Mean_F1_Score'] = np.mean(list(f1_scores.values()))
    return f1_scores


def mean_absolute_error_multiclass(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def hausdorff_distance_multiclass(y_true, y_pred, num_classes):
    hausdorff_scores = {}
    for i in range(num_classes):
        y_true_points = np.argwhere(y_true[..., i] > 0)
        y_pred_points = np.argwhere(y_pred[..., i] > 0)
        forward_hd = directed_hausdorff(y_true_points, y_pred_points)[0]
        backward_hd = directed_hausdorff(y_pred_points, y_true_points)[0]
        hausdorff_scores[f'Class_{i}'] = max(forward_hd, backward_hd)
    return hausdorff_scores
```

### Percentage

```python
def calculate_class_percentages(mask):
    """
    Calculate the percentage of each class in a segmentation mask.

    Parameters:
    mask (numpy.ndarray): A 3D array representing the RGB mask with shape (256, 256, 3).
                          Red channel (R): Deforested area
                          Green channel (G): Forest cover
                          Blue channel (B): Other areas

    Returns:
    None: Prints the percentage of each class with appropriate messages.
    """
    # Validate the input mask
    if mask.shape != (256, 256, 3):
        raise ValueError("The input mask must have dimensions (256, 256, 3).")

    # Calculate total number of pixels
    total_pixels = mask.shape[0] * mask.shape[1]

    # Determine class masks
    deforested_mask = (mask[:, :, 0] >= RED[0]) 
    forest_mask = (mask[:, :, 1] >= GREEN[1]) 
    other_mask = (mask[:, :, 2] >= BLUE[2]) 

    # Calculate pixel counts for each class
    deforested_count = np.sum(deforested_mask)
    forest_count = np.sum(forest_mask)
    other_count = np.sum(other_mask)

    # Compute percentages
    deforested_percentage = (deforested_count / total_pixels) * 100
    forest_percentage = (forest_count / total_pixels) * 100
    other_percentage = (other_count / total_pixels) * 100

    # Print results
    print(f"Deforested area: {deforested_percentage:.2f}% of the region.")
    print(f"Forest cover: {forest_percentage:.2f}% of the region.")
    print(f"Other areas: {other_percentage:.2f}% of the region.")
```
